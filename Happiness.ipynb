```python
import pandas as pd
import glob
```

```python
file="2015.csv"
data= pd.read_csv(file)
data.head()
```

```python
len(data)
```

```python
countries_2015 = data["Country"].count()
countries_2015
```

```python
data["Year"]= 2015
data.head()
```

```python
file2="2016.csv"
data2= pd.read_csv(file)
data2.head()
```

```python
len(data2)
```

```python
countries_2016 = data2["Country"].count()
countries_2016
```

```python
data2["Year"]= 2016
data2.head()
```

```python
data2015= data[["Country","Happiness Score"]].copy()
data2015= data2015.rename(columns={"Happiness Score":"Score 2015"})
data2016= data2[["Country","Happiness Score"]].copy()
data2016= data2016.rename(columns={"Happiness Score":"Score 2016"})
data2016.head()
```

```python
combined_data = pd.merge(data2015, data2016, how='inner', on='Country')
combined_data.head()
```
